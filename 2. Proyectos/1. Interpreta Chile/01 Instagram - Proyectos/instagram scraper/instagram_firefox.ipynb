```python
def ManagerInstagramScrapper(urls, medio, username, password):

    # Configuraci贸n Firefox
    firefox_options = Options()
    driver = webdriver.Firefox(options=firefox_options)
    # driver.maximize_window()

    # LOGIN UNA SOLA VEZ
    driver.get("https://www.instagram.com/")
    time.sleep(3)

    try:
        username_input = driver.find_element(By.CSS_SELECTOR, "input[aria-label='Tel茅fono, usuario o correo electr贸nico']")
        password_input = driver.find_element(By.XPATH, "//input[@aria-label='Contrase帽a']")
    except:
        username_input = driver.find_element(By.CSS_SELECTOR, "input[aria-label='N煤mero de m贸vil, nombre de usuario o correo electr贸nico']")
        password_input = driver.find_element(By.XPATH, "//input[@aria-label='Contrase帽a']")

    username_input.send_keys(username)
    password_input.send_keys(password)
    password_input.send_keys(Keys.RETURN)
    time.sleep(10)  # Esperar login

    for url in urls:
        driver.get(url)
        time.sleep(5)

        # Scroll para mostrar comentarios
        pyautogui.moveTo(465, 465, duration=0.3)
        pyautogui.click()
        time.sleep(1)

        try:
            pyautogui.moveTo(1040, 283, duration=0.3)
            pyautogui.click()
            for _ in range(60):
                pyautogui.scroll(-305)
                time.sleep(2.4)
        except:
            pass

        
        try:
            print('abriendo paneles de respuesta')
            
            driver.execute_script("""    
                Array.from(document.querySelectorAll('div.x1i10hfl.xjbqb8w.x1ejq31n.x18oe1m7'))
                .filter(el => el.innerText.includes("Ver las"))
                .forEach(el => el.click());
            """)
            time.sleep(5)

            print('mostrando mas respuestas')
            driver.execute_script("""    
                Array.from(document.querySelectorAll('div.x1lliihq.x1plvlek.xryxfnj.x1n2onr6'))
                .filter(el => el.innerText.includes("Mostrar"))
                .forEach(el => el.click());
            """)
            
            time.sleep(3)
            print('Creando Excel')

        except:
            pass 


        # Extraer comentarios principales
        comentarios_principales = driver.find_elements(By.CSS_SELECTOR, "div.xdj266r.x14z9mp.xat24cr.x1lziwak" ) #se necesita buscar el bloque correcto, de solo usuario y comentario
        textos_comentarios_principales = [comentario.text for comentario in comentarios_principales]

        resultado_principal = extraer_ultimos_dos_por_bloque(textos_comentarios_principales)
        
        df_principal = pd.DataFrame(resultado_principal, columns=["Usuario", "Comentario"]).drop_duplicates(subset=["Usuario", "Comentario"])
        df_principal["url_instagram"] = url

        df = df_principal


        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f'data/data_{medio}_{timestamp}.xlsx'
        df.to_excel(filename, index=False)

        # filename = f'data/data_{medio}_{timestamp}.csv'
        # df.to_csv(filename, index=False,encoding="utf-8-sig")

        print(f"Excel generado para {url}")
        


    driver.quit()
    print("Navegador cerrado y proceso finalizado.")
    
    return textos_comentarios_principales

```
