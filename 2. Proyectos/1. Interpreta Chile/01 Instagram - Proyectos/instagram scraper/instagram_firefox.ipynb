```python
def ManagerInstagramScrapper(urls, medio, username, password):

    # Configuración Firefox
    firefox_options = Options()
    driver = webdriver.Firefox(options=firefox_options)
    # driver.maximize_window()

    # LOGIN UNA SOLA VEZ
    driver.get("https://www.instagram.com/")
    time.sleep(3)

    try:
        username_input = driver.find_element(By.CSS_SELECTOR, "input[aria-label='Teléfono, usuario o correo electrónico']")
        password_input = driver.find_element(By.XPATH, "//input[@aria-label='Contraseña']")
    except:
        username_input = driver.find_element(By.CSS_SELECTOR, "input[aria-label='Número de móvil, nombre de usuario o correo electrónico']")
        password_input = driver.find_element(By.XPATH, "//input[@aria-label='Contraseña']")

    username_input.send_keys(username)
    password_input.send_keys(password)
    password_input.send_keys(Keys.RETURN)
    time.sleep(8)  # Esperar login

    for url in urls:
        driver.get(url)
        time.sleep(5)

        # Scroll para mostrar comentarios
        pyautogui.moveTo(465, 465, duration=1)
        pyautogui.click()
        time.sleep(1)

        try:
            pyautogui.moveTo(938, 286, duration=1)
            pyautogui.click()
            for _ in range(10):
                pyautogui.scroll(-300)
                time.sleep(2.5)
        except:
            pass

        # Extraer comentarios

        comentarios = driver.find_elements(By.CSS_SELECTOR, "div.x78zum5.xdt5ytf.x1iyjqo2") #se necesita buscar el bloque correcto, de solo usuario y comentario
        textos_comentarios = [comentario.text for comentario in comentarios]

        # Procesar comentarios con tu función modelamiento
        # procesados = [modelamiento(texto) for texto in textos_comentarios]
        # flattened_data = [item for sublist in procesados for item in sublist]

        resultado = limpiar_comentarios_(textos_comentarios)
        

        # df = pd.DataFrame(flattened_data, columns=["Usuario", "Comentario", "Hora", "Reacciones"])
        df = pd.DataFrame(resultado, columns=["Usuario", "Comentario","Likes"]).drop_duplicates(subset=["Usuario", "Comentario"])
        
        df['Comentario'] = [x.split(' ', 2)[2] if len(x.split(' ', 2)) > 2 else '' for x in df['Comentario']]

        
        df["url_instagram"] = url

        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f'data/data_{medio}_{timestamp}.xlsx'
        df.to_excel(filename, index=False)

        print(f"Excel generado para {url}")
        # print(resultado)


    driver.quit()
    print("Navegador cerrado y proceso finalizado.")

    return textos_comentarios
```
