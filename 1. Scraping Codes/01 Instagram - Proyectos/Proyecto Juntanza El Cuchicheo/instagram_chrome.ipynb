```python
#VERSION 02

#VARIABLES PARA EL SCRAPING
comentarios_recorridos = list()
centinela = True
comment_interno = 1
comment_externo = 1
comment_respuestas = 1
comment_total_respuestas = 3

base_final = pd.DataFrame()

while centinela == True:
    try:
        
    ##VALIDACION DE BLOQUE DE COMENTARIOS           
        try:
            driver.find_element(By.XPATH, value= '/html/body/div[2]/div/div/div[2]/div/div/div/div[1]/div[1]/div[2]/section/main/div/div[1]/div/div[2]/div/div[2]/div/div[1 ]/div/div[2]/div/span/div/div/span[1]/div/a/div/div/span')
            print('Usuario Encontrado:')
            time.sleep(3)
        except:
            pass

    ## PRIMER NIVEL DE COMENTARIOS
        # Level 1 comment principal
        level = 1
        # Detectar el DISPLAY NAME del usuario
        try:
            display_name = driver.find_element(By.XPATH, value = seccion_comments + str(comment_externo) + "]/div/div[2]/div/span/div/div/span[1]/div/a").text
            
            username_thread = driver.find_element(by=By.PARTIAL_LINK_TEXT,value=display_name).get_attribute("href")
            comment = driver.find_element(By.XPATH, value = seccion_comments + str(comment_externo) + "]/div/div[2]/div/span/div/span").text

            print(display_name + '_bloque01_' + username_thread + comment)

        except:
            try:
                display_name = driver.find_element(By.XPATH, value = seccion_comments + str(comment_externo) + "]/div/div/div/div[2]/div[1]/div[1]/div/div[1]/span[1]/span/div/a/div/div/span").text

                username_thread = driver.find_element(by=By.PARTIAL_LINK_TEXT,value=display_name).get_attribute("href")
                comment = driver.find_element(By.XPATH, value = seccion_comments + str(comment_externo) + "/div/div/div/div[2]/div[1]/div[1]/div/div[2]/span").text
                
                print(display_name + '_bloque04_' + username_thread + comment)                                          

        
            except:

                display_name = driver.find_element(By.XPATH, value = seccion_comments + str(2) +"]/div[" + str(comment_interno) + "]/div/div/div[2]/div[1]/div[1]/div/div[1]/span[1]/span/div/a/div/div/span").text

                username_thread = driver.find_element(by=By.PARTIAL_LINK_TEXT,value=display_name).get_attribute("href")
                comment = driver.find_element(By.XPATH, value = seccion_comments + str(2) +"]/div[" + str(comment_interno) + "]/div/div/div[2]/div[1]/div[1]/div/div[2]/span").text
                
                print(display_name + '_bloque02_' + username_thread + comment)


                comment_interno  = comment_interno + 1
                try:
                    '''Click Ver Respuesta'''
                    ver_respuesta = driver.find_element(By.XPATH, value = seccion_sub_comments + str(comment_respuestas) + "]/div[2]/div/div/span")
                    ver_respuesta.click()
                    comment_respuestas  = comment_respuestas + 1
                    print('click realizado', comment_respuestas)
                    time.sleep(5)   

                except: 
                    print("no_click")
                    pass
                

        # except: 
        #     pass
        

        
    ##DATAFRAME CON COMENTARIOS
        # Creacion de Dataframe con los datos de cada comentario, y la base previa 
        d = {"level": level,"Instragram Post":INSTAGRAM_URL, "display_name": display_name, "username":username_thread, "comment":comment}
        base_temporal = pd.DataFrame(data=d,index=[comment_externo])
        base_final = pd.concat([base_final, base_temporal], ignore_index=True)
    

    ##LOOP
        # Actualizacion del loop para continuar con el recorrido
        comentarios_recorridos.append(comment_externo)
        comment_externo  = comment_externo + 1
        
        
        print(comment_externo)
        print("Acabé el comment " + str(comment_externo - 1) + ' perteneciente al usuario: ' + str(display_name) + '\n')
    
        time.sleep(5)
        
    except :
        break
        

print('CARGA COMPLETA')

# driver.quit()

base_final.to_excel('base_instagram_mutante_23.xlsx')
```
