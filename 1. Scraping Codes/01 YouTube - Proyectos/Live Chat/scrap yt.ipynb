```python
import pytchat
import pandas as pd
```

```python
# ID del video en vivo de YouTube
video_id = "W7TuY1R1tyg"  # Reemplaza con el ID real

# Crear objeto del chat
chat = pytchat.create(video_id=video_id)

# DataFrame vacío para acumular mensajes
df_chat = pd.DataFrame(columns=["autor", "mensaje", "timestamp"])

print("⏳ Capturando mensajes... (Ctrl+C para detener)")

try:
    while chat.is_alive():
        nuevos = []
        for c in chat.get().sync_items():
            timestamp_legible = pd.to_datetime(int(c.timestamp), unit='ms')
            mensaje = {
                "autor": c.author.name,
                "mensaje": c.message,
                "timestamp": timestamp_legible
            }
            nuevos.append(mensaje)

            # Mostrar en consola
            print(f"[{mensaje['timestamp']}] {mensaje['autor']}: {mensaje['mensaje']}")

        # Agregar nuevos mensajes al DataFrame
        if nuevos:
            df_nuevos = pd.DataFrame(nuevos)
            df_chat = pd.concat([df_chat, df_nuevos], ignore_index=True)

except KeyboardInterrupt:
    print("\n🛑 Captura detenida manualmente.")

# Mostrar resumen al final
print(f"\n✅ Total de mensajes capturados: {len(df_chat)}")
```

```python
df_chat.to_excel('live_chat.xlsx',index=False)
```
